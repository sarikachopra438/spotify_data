# EXPLORATORY DATA  ANALYSIS

**Which artist or musician has the most number of followers on spotify?**

```python
sort_df=df.sort_values('popularity',ascending=False)
sort_df.head(10)
```

**Top 10 most popular songs each are less than 90 & greater than 80**

```python
pop_df=df.loc[(df['popularity']<90) & (df['popularity']>=80)].sort_values('popularity', ascending=False)
pop_df[:10]
```

**Check the artist at the 18th row in our datset**

```python
ef=df[['artists']].iloc[18]
ef
```

**Covert the duration in MS to Seconds**

```python
df['duration']=df['duration_ms'].apply(lambda x: round(x/1000))
df.drop('duration_ms',inplace=True,axis=1)
df.head()
```

**Drop three unwanted columns and Those are 'key' , 'mode' , 'explicit'**

```python
drop_df=df.drop(['key','mode','explicit'],axis=1)
```

**Show the dataset in heatmap** 

```python
L=list(df.columns)
```

```python
plt.figure(figsize=(14,6))
heatmap=sns.heatmap(drop_df.corr(), cmap="crest", annot=True)
heatmap.set_title('Correlation HeatMap Between Variable ')
heatmap.set_xticklabels(heatmap.get_xticklabels(),rotation=90)
plt.show()
```

**Sample at data and take just 0.4% of the total data and will create the regression plot using this dataset**

```python
sample_df=df.sample(int(0.004*len(df)))
len(sample_df)
```

**Create regression plot between loudness & energy**

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df,y='loudness',x='energy',color='c').set(title='Loudness VS Energy Correlation')
```

There is a very high positive correlation between loudness and energy.

**See another Regression plot this time for two diffrent features**

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df,y='popularity',x='acousticness',color='b').set(title='popularity VS acousticness Correlation')
```

**Creat a new column called year from our realese_date column**

```python
df['year']=df['release_date'].dt.year
df['year'].head()
```

**Creat a distribution plot to visualize the total no of songs and each years since 1922 that is available on the spotify streaming app**

```python
sns.displot(df['year'],discrete=True,aspect=2,height=5,kind='hist',).set(title='Number Of Songs Per Year')
```

**See the duration of songs over the years for that again we are going to creat a bar plot**

```python
total_dr=df.duration
years=df.year
fig, ax = plt.subplots(figsize=(18,7))
fig=sns.barplot(x=years,y=total_dr,ax=ax,errwidth=False).set(title='YEAR vs DURATION')
plt.xticks(rotation=90)
```

**Creat a line plot to analyze the average duration of the song over the years**

```python
total_dr=df.duration
sns.set_style(style='whitegrid')
years=df.year
fig, ax = plt.subplots(figsize=(18,7))
fig=sns.lineplot(x=years,y=total_dr,ax=ax).set(title='YEAR vs DURATION')
plt.xticks(rotation=60)
```
